```python
from ipywidgets import interact,interact_manual
```

```python
num_col = df.select_dtypes(include=['int64', 'float64']).columns
```

```python
def outliers(data):
    for col in num_col:
        q1=data[col].quantile(0.25)
        q3=data[col].quantile(0.75)
        iqr=q3-q1
        lb=q1-(1.5*iqr)
        ub=q3+(1.5*iqr)
        olr=list(data[(data[col]>=ub)|(data[col]<=lb)][col])
        print(f"Number of outliers in column '{col}': {len(olr)}")
outliers(df)
```

# Average Salary of Students

```python
import matplotlib.pyplot as plt

# Calculate the average salary
avg_salary = df['Salary'].mean()

# Filter salaries below and above the average
below_avg = df[df['Salary'] < avg_salary]['Salary']
above_avg = df[df['Salary'] > avg_salary]['Salary']

# Plot histogram
plt.hist(below_avg, color='blue', alpha=0.5, label='Below Average Salary')
plt.hist(above_avg, color='red', alpha=0.5, label='Above Average Salary')
plt.axvline(avg_salary, color='black', linestyle='dashed', linewidth=1, label='Average Salary')
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

>* It is a right skewed distribution. Most of the graduate students are getting below average salaries
